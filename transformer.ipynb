```python
from datasets import load_dataset

# Load the dataset
dataset = load_dataset("ai4privacy/pii-masking-400k")

# Filter for rows where language is 'en'
filtered_dataset = dataset['train'].filter(lambda example: example['language'] == 'en')

# Select only the required columns
filtered_dataset = filtered_dataset.remove_columns([col for col in filtered_dataset.column_names if col not in ['mbert_tokens', 'mbert_token_classes']])

# Inspect the filtered dataset
print(filtered_dataset[0])  # Print the first example
```

```python
# Extract all unique tokens from mbert_tokens
def create_vocabulary(dataset, special_tokens=None):
    # Flatten the list of tokens from all rows
    all_tokens = [token for example in dataset for token in example['mbert_tokens']]
    unique_tokens = set(all_tokens)  # Get unique tokens

    # Create a token-to-ID mapping
    vocab = {token: idx for idx, token in enumerate(unique_tokens)}

    # Add special tokens
    if special_tokens:
        for special_token in special_tokens:
            if special_token not in vocab:
                vocab[special_token] = len(vocab)

    return vocab

# Create vocabulary with special tokens
special_tokens = ["[PAD]", "[UNK]"]
vocab = create_vocabulary(filtered_dataset, special_tokens=special_tokens)

# Print vocabulary size and sample
print(f"Vocabulary size: {len(vocab)}")
print(f"Sample tokens: {list(vocab.keys())[:10]}")
```

```python
# Convert tokens in the dataset to token IDs using the created vocabulary
def tokens_to_ids(example):
    token_ids = [vocab.get(token, vocab["[UNK]"]) for token in example['mbert_tokens']]
    return {"input_ids": token_ids}

# Apply the conversion to the filtered dataset
filtered_dataset = filtered_dataset.map(tokens_to_ids)

# Inspect the dataset
print(filtered_dataset[0])
```

```python
# Define the max sequence length threshold
MAX_LENGTH_THRESHOLD = 128

# Filter the dataset by sequence length
filtered_dataset = filtered_dataset.filter(lambda example: len(example['input_ids']) <= MAX_LENGTH_THRESHOLD)

# Check the first few entries of the filtered dataset
print(filtered_dataset[0])
```

```python
MAX_SEQ_LEN = 128
PAD_TOKEN_ID = vocab["[PAD]"]  # Use the `[PAD]` token from your vocabulary

def preprocess_function(example):
    # Pad or truncate input_ids
    input_ids = example['input_ids'][:MAX_SEQ_LEN]
    seq_len = len(input_ids)
    if seq_len < MAX_SEQ_LEN:
        input_ids += [PAD_TOKEN_ID] * (MAX_SEQ_LEN - seq_len)

    # Create attention mask
    attention_mask = [1] * seq_len + [0] * (MAX_SEQ_LEN - seq_len)
    attention_mask = attention_mask[:MAX_SEQ_LEN]

    # Pad or truncate labels
    labels = example['mbert_token_classes'][:MAX_SEQ_LEN]
    labels += ["O"] * (MAX_SEQ_LEN - len(labels))

    return {
        "input_ids": input_ids,
        "attention_mask": attention_mask,
        "labels": labels
    }

processed_dataset = filtered_dataset.map(preprocess_function)
print(processed_dataset[0])  # Inspect preprocessed data
```

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.functional import softmax

# Encoder Components
class MultiHeadAttention(nn.Module):
    def __init__(self, d_model, num_heads):
        super(MultiHeadAttention, self).__init__()
        assert d_model % num_heads == 0
        self.num_heads = num_heads
        self.depth = d_model // num_heads

        # Define the layers
        self.wq = nn.Linear(d_model, d_model)
        self.wk = nn.Linear(d_model, d_model)
        self.wv = nn.Linear(d_model, d_model)
        self.dense = nn.Linear(d_model, d_model)

    def split_heads(self, x, b\atch_size):
        """Split the last dimension into (num_heads, depth)."""
        x = x.view(batch_size, -1, self.num_heads, self.depth)
        return x.permute(0, 2, 1, 3)  # (batch_size, num_heads, seq_len, depth)

    def forward(self, x, mask=None):
        batch_size = x.size(0)

        # Generate query, key, value tensors
        query = self.wq(x)  # (batch_size, seq_len, d_model)
        key = self.wk(x)
        value = self.wv(x)

        # Split into heads
        query = self.split_heads(query, batch_size)
        key = self.split_heads(key, batch_size)
        value = self.split_heads(value, batch_size)

        # Scaled dot-product attention
        d_k = query.size(-1)
        scores = torch.matmul(query, key.transpose(-2, -1)) / torch.sqrt(torch.tensor(d_k, dtype=torch.float32))  # (batch_size, num_heads, seq_len, seq_len)

        if mask is not None:

            mask = mask.unsqueeze(1).unsqueeze(1)
            scores += (mask * -1e9)

        # Apply softmax to get attention weights
        attention_weights = torch.nn.functional.softmax(scores, dim=-1)  # (batch_size, num_heads, seq_len, seq_len)

        # Apply attention weights to value tensor
        attention_output = torch.matmul(attention_weights, value)  # (batch_size, num_heads, seq_len, depth)

        # Combine the heads back
        attention_output = attention_output.permute(0, 2, 1, 3).contiguous().view(batch_size, -1, self.num_heads * self.depth)

        # Final dense layer
        output = self.dense(attention_output)  # (batch_size, seq_len, d_model)
        return output

class FeedForward(nn.Module):
    def __init__(self, d_model, d_ff):
        super().__init__()
        self.linear1 = nn.Linear(d_model, d_ff)
        self.linear2 = nn.Linear(d_ff, d_model)

    def forward(self, x):
        x = self.linear1(x)
        x = torch.relu(x)
        x = self.linear2(x)
        return x

class LayerNorm(nn.Module):
    def __init__(self, d_model, epsilon=1e-6):
        super().__init__()
        self.gamma = nn.Parameter(torch.ones(d_model))
        self.beta = nn.Parameter(torch.zeros(d_model))
        self.epsilon = epsilon

    def forward(self, x):
        mean = x.mean(-1, keepdim=True)
        variance = x.var(-1, keepdim=True, unbiased=False)
        normalized = (x - mean) / torch.sqrt(variance + self.epsilon)
        return self.gamma * normalized + self.beta

class EncoderLayer(nn.Module):
    def __init__(self, d_model, num_heads, d_ff):
        super().__init__()
        self.mha = MultiHeadAttention(d_model, num_heads)
        self.ffn = FeedForward(d_model, d_ff)
        self.ln1 = LayerNorm(d_model)
        self.ln2 = LayerNorm(d_model)

    def forward(self, x, mask=None):
        attn_output = self.mha.forward(x, mask)
        x = self.ln1(x + attn_output)
        ffn_output = self.ffn(x)
        x = self.ln2(x + ffn_output)
        return x

class Encoder(nn.Module):
    def __init__(self, num_layers, d_model, num_heads, d_ff, input_vocab_size, max_seq_len):
        super().__init__()
        self.embedding = nn.Embedding(input_vocab_size, d_model)
        self.pos_encoding = self.positional_encoding(max_seq_len, d_model)
        self.enc_layers = nn.ModuleList([EncoderLayer(d_model, num_heads, d_ff) for _ in range(num_layers)])
        self.d_model = d_model

    def positional_encoding(self, max_seq_len, d_model):
        pos = np.arange(max_seq_len)[:, np.newaxis]
        i = np.arange(d_model)[np.newaxis, :]
        angle_rates = 1 / np.power(10000, (2 * (i // 2)) / np.float32(d_model))
        pos_encoding = pos * angle_rates
        pos_encoding[:, 0::2] = np.sin(pos_encoding[:, 0::2])
        pos_encoding[:, 1::2] = np.cos(pos_encoding[:, 1::2])
        return torch.tensor(pos_encoding, dtype=torch.float32)

    def forward(self, x, mask=None):
        seq_len = x.size(1)
        x = self.embedding(x) * np.sqrt(self.d_model)
        x += self.pos_encoding[:seq_len, :].to(x.device)

        for layer in self.enc_layers:
            x = layer(x, mask)
        return x

class SequenceClassifier(nn.Module):
    def __init__(self, encoder, num_classes):
        super(SequenceClassifier, self).__init__()
        self.encoder = encoder
        self.fc = nn.Linear(encoder.d_model, num_classes)  # Final output layer for classification

    def forward(self, input_ids, attention_mask=None):
        # Forward pass through the encoder (assumes attention_mask is handled inside the encoder)
        encoder_output = self.encoder(input_ids, attention_mask)


        sequence_output = encoder_output

        # Apply the final classification layer
        logits = self.fc(sequence_output)
        return logits
```

```python
# Training Code
def train_model_with_processed_dataset(model, processed_dataset, num_epochs, batch_size, learning_rate, device):
    model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    loss_fn = nn.CrossEntropyLoss()

    input_ids = torch.tensor(processed_dataset['input_ids'], dtype=torch.long).to(device)
    attention_mask = torch.tensor(processed_dataset['attention_mask'], dtype=torch.float32).to(device)
    labels = torch.tensor(processed_dataset['labels'], dtype=torch.long).to(device)

    dataset_size = input_ids.size(0)

    for epoch in range(num_epochs):
        epoch_loss = 0
        permutation = torch.randperm(dataset_size)

        for batch_idx in range(0, dataset_size, batch_size):
            indices = permutation[batch_idx: batch_idx + batch_size]
            batch_input_ids = input_ids[indices]
            batch_attention_mask = attention_mask[indices]
            batch_labels = labels[indices]

            optimizer.zero_grad()
            predictions = model(batch_input_ids, batch_attention_mask)
            loss = loss_fn(predictions.view(-1, predictions.size(-1)), batch_labels.view(-1))
            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss / (dataset_size // batch_size)}")

# Hyperparameters
import torch
from torch.utils.data import DataLoader, TensorDataset

# Hyperparameters
device = torch.device("cpu")  # Force CPU
input_vocab_size = 24030
max_seq_len = 128
num_classes = 35
num_layers = 2
d_model = 128
num_heads = 4
d_ff = 512
num_epochs = 15
batch_size = 32
learning_rate = 0.001

# Create a DataLoader from processed_dataset
def create_dataloader(processed_dataset, batch_size):
    input_ids = torch.tensor(processed_dataset['input_ids'], dtype=torch.long)
    attention_mask = torch.tensor(processed_dataset['attention_mask'], dtype=torch.float32)
    labels = torch.tensor(processed_dataset['labels'], dtype=torch.long)
    dataset = TensorDataset(input_ids, attention_mask, labels)
    return DataLoader(dataset, batch_size=batch_size, shuffle=True)


# Adjusted training loop with gradient accumulation
def train_model_on_cpu(model, dataloader, num_epochs, learning_rate, device, accumulation_steps):
    model.to(device)  # Ensure the model is on the correct device (CPU)
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    loss_fn = torch.nn.CrossEntropyLoss()

    for epoch in range(num_epochs):
        epoch_loss = 0
        model.train()

        # Accumulate gradients over multiple mini-batches
        optimizer.zero_grad()  # Ensure no gradients are left over from previous iteration

        for i, (batch_input_ids, batch_attention_mask, batch_labels) in enumerate(dataloader):
            # Move data to CPU device
            batch_input_ids, batch_attention_mask, batch_labels = (
                batch_input_ids.to(device),
                batch_attention_mask.to(device),
                batch_labels.to(device),
            )

            # Forward pass
            output = model(batch_input_ids, attention_mask=batch_attention_mask)
            loss = loss_fn(output.view(-1, num_classes), batch_labels.view(-1))

            # Backward pass
            loss.backward()

            if (i + 1) % accumulation_steps == 0:
                optimizer.step()
                optimizer.zero_grad()

            epoch_loss += loss.item()


        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss / len(dataloader)}")

encoder = Encoder(num_layers, d_model, num_heads, d_ff, input_vocab_size, max_seq_len)
model = SequenceClassifier(encoder, num_classes)


device = torch.device('cpu')


accumulation_steps = 4


dataloader = create_dataloader(processed_dataset, batch_size)
train_model_on_cpu(model, dataloader, num_epochs, learning_rate, device, accumulation_steps)
```
